```python
def resize_image(image_data, target_size=(20, 20)):

  if len(image_data.shape) != 1:
    raise ValueError("Datapoint should be a 1D array representing the flattened image.")

  # Reshape the data to 2D image format based on original size (assuming 28x28)
  image_size = 28  # Adjust this if your original images have a different size
  reshaped_image = image_data.reshape(image_size, image_size)

  # Convert to PIL Image and resize using target dimensions
  image = Image.fromarray(reshaped_image.astype(np.uint8))
  resized_image = image.resize(target_size, Image.ANTIALIAS)

  # Convert the resized image back to a NumPy array
  resized_data = np.array(resized_image)
  resized_data = resized_data.flatten()

  return resized_data
```

# **Splitting the dataframe into input and outut. And splitting into train and test with 80:20 split**

```python
y=df_final["label"]
X=df_final.drop("label",axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.80,random_state=12)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# **Applying Transformations**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_transformed=pd.DataFrame(scaler.fit_transform(X_train),index=X_train.index,columns=X_train.columns)
X_train_transformed.head()
```

```python
X_test_transformed=pd.DataFrame(scaler.transform(X_test),index=X_test.index,columns=X_test.columns)
X_test_transformed.head()
```

# **Model Building and Prediction**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
```

```python
KNN = KNeighborsClassifier()
LR = LogisticRegression(max_iter=5000)
SVC = SVC()
DT = DecisionTreeClassifier()
RF = RandomForestClassifier()
GDB = GradientBoostingClassifier()
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(X_train, y_train)
y_pred = NB.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print('Model Accuracy:',accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(X_train_transformed, y_train)
y_pred = NB.predict(X_test_transformed)
NB_res = metrics.accuracy_score(y_test, y_pred)
print('Model Accuracy:',NB_res)
```
